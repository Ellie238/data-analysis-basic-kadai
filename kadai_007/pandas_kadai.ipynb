```python
# matplotlib.pyplot、pandas、japanize_matplotlibpandasをインポートする
import matplotlib.pyplot as plt
import pandas as pd
!pip install japanize-matplotlib
import japanize_matplotlib

# csvを読み込む
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv("category.csv")

# マージする
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# カテゴリー列の要素の出現頻度をカウントする
category_counts = df['カテゴリー'].value_counts()

# 棒グラフを作成する
category_counts.plot(kind='bar')
plt.title('カテゴリー別の出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.show()

# 商品番号ごとに注文数の基本統計量を求める
order_summary = df.groupby('商品番号')['注文数'].describe()
print(order_summary)
```
